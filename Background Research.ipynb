#### Sensual Impressions

The Yoga Sutras are accepted by Vedic schools as the most authoritative source on Yoga teachings.

Yoga is defined by Patanjali as "the stilling of all states in the citta". There are five sensual impressions (thoughts/ideas/conscious mental states): right knowledge, error, metaphor, deep sleep, and memory. Each is either detrimental or conductive to the goals of Yoga.

##### I. Pramana 

Sources that constitute the production of valid knowledge or when the mind is in a state of right knowledge (accurately reflecting external reality).

##### II. Error

Considering something to be what it is not. This can be subsequently removed by true knowledge of what the nature of a thing in question is

##### III. Imaginataion (Metaphor)

Use of words or expressions that don't correspond to any actual physical reality but that are understood by most.

For example "the arrow stands still, stood still, will stand still"

the absence of motion, is really an imagined state of affairs dependent on the idea of motion, but it is then projected as an actual characteristic of the arrow

##### IV. Deep Sleep

State of mind based on an absence of any content. Generally viewed that when one awakes, they remember if they slept well, restlessly, or in a stupor. This is only possible because these impressions relate back to a state of mind that existed during deep sleep. In Yoga psychology, memory is the product of samskara and that is caused by experience. 

###### V. Memory

The retention of images of sense objects that have been experienced. Memory is generated by and dependent on the other four sensual impressions.

#### The Eight Limbs

- Yoga is the progressive settling of the mind into silence
- Our essential nature is usually overshadowed by the activity of the mind
- When the mind is settled, we are established in our own essential state, which is unbounded consciousness

Essentially, spiritual practice should be looking within ones self. 



##### I. Yama

Correct behavior towards others (nonviolence, truthfulness, not stealing, not wasting energy, abstaining from greed)

##### II. Niyama

The principles of how to live your own life (purity, contentment, study, devotion, spiritual observances)

##### III. Asana

The seat of consciousness (postures to prepare the body)

##### IV. Pranayama

How to expand life force via breathing exercise

##### V. Pratyahara

Turn the senses inwards to explore inner universe

###### VI. Dharana

Effortless focused attention (training mind to meditate)


###### VII. Dhyana

A continous flow (perfected meditation)

##### VIII. Samadhi 

Lost or found in the Divine (unity)

### Upanishads

Country of Origin: India

Time Period Written: ~800BC or later

Attributed Writter: Vyasa

Associated Religion: Hinduism

The central theme is the idea of liberataing the soul and returning to the world of Brahman.

##### I. Karma

PAying for one's actions or sins. An individual's intents and actions will endure consequences both good and bad. Also affects their next life.

##### II. Samsara

Reincarnaton.

##### III. Dharma

Duty. Everyone has a set of responsibilities in accordance to their caste.

##### IV. Moksha

Release fro the cycle of rebirth, the ultimate aim of an individual's mind and soul --> individual breaks away from endless cycle and finds solace
